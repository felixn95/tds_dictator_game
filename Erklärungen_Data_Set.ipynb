### Erklärung der Binarisierung des Diktatorspiels

Das traditionelle Diktatorspiel besteht darin, dass ein Spieler (der Diktator) einen bestimmten Geldbetrag erhält und diesen zwischen sich und einem anderen Spieler (dem Empfänger) aufteilen muss. Der Diktator hat die volle Kontrolle über die Aufteilung, und der Empfänger muss die Entscheidung akzeptieren.

In der hier verwendeten binären Version des Diktatorspiels gibt es jedoch eine festgelegte Struktur: Der Diktator hat nur zwei feste Optionen (X und Y), zwischen denen er wählen kann. Jede Option gibt vor, wie viel der Diktator und der Empfänger erhalten. Diese Binarisierung vereinfacht das Entscheidungsproblem und ermöglicht es den Forschern, spezifische Präferenzen und Verhaltensmuster leichter zu analysieren.

### Allgemeine Beschreibung der Datensätze und deren Zweck

Die bereitgestellten Datensätze stammen aus einer experimentellen Studie, die das Entscheidungsverhalten von Individuen im sogenannten Diktatorspiel untersucht. Ziel des Experiments ist es zu verstehen, wie Personen (Diktatoren) Ressourcen zwischen sich selbst und anderen (Empfängern) verteilen und welche Faktoren diese Entscheidungen beeinflussen könnten.

### Beschreibung der Datensätze

1. **choices.csv**: Dieser Datensatz erfasst die Entscheidungen der Diktatoren im binären Diktatorspiel über 30 verschiedene Varianten. Jede Zeile stellt eine einzelne Entscheidung eines Subjekts in einer bestimmten Spielvariante dar.

    - *sid*: ID des Subjekts.
    - *gid*: ID der Spielvariante.
    - *order*: Die zufällige Reihenfolge, in der das Spiel in der Sequenz der Spiele erschien.
    - *self_x* und *self_y*: Die Auszahlungen für den Diktator in den Optionen X und Y.
    - *other_x* und *other_y*: Die Auszahlungen für den Empfänger in den Optionen X und Y.
    - *choice_x*: Gibt die Entscheidung des Diktators an (1, wenn Option X gewählt wurde, 0, wenn Option Y gewählt wurde).
    - *s_x* und *s_y*: Indikatoren für nachteilige Ungleichheit für den Diktator in den Optionen X und Y.
    - *r_x* und *r_y*: Indikatoren für vorteilhafte Ungleichheit für den Diktator in den Optionen X und Y.

2. **subjects.csv**: Dieser Datensatz enthält Informationen über die Subjekte, einschließlich ihrer Persönlichkeitseigenschaften, kognitiven Fähigkeiten, Studienrichtung, monatlichem Einkommen, Alter und Geschlecht.

    - *sid*: ID des Subjekts.
    - *bf_consciousness, bf_openness, bf_extraversion, bf_agreeableness, bf_neuroticism*: Ergebnisse der Big Five-Persönlichkeitsmerkmale.
    - *cogabil*: Maß für kognitive Fähigkeiten.
    - *pe_D1_stud_natsci, pe_D1_stud_law, pe_D1_stud_socsci, pe_D1_stud_med*: Indikatoren für das Studienfach des Subjekts (1, wenn sie in diesem Fach studieren, 0 ansonsten).
    - *pe_monthinc*: Monatliches Einkommen.
    - *pe_age*: Alter des Subjekts.
    - *pe_female*: Geschlecht des Subjekts (1, wenn weiblich, 0 ansonsten).

### Zusammenführung der Datensätze

Die beiden Datensätze können über die *sid*-Spalte, die in beiden Datensätzen vorhanden ist, miteinander verknüpft werden. Dadurch können die Entscheidungen der Diktatoren im Spiel (choices.csv) mit ihren demografischen und psychologischen Merkmalen (subjects.csv) verknüpft werden.

### Zweck und Analyse

Durch die Kombination dieser Datensätze können wir analysieren, wie verschiedene Faktoren das Entscheidungsverhalten der Diktatoren beeinflussen. Beispielsweise können wir untersuchen:

1. **Persönlichkeitseigenschaften**: Wie korrelieren die Big Five-Persönlichkeitsmerkmale mit den Entscheidungen der Diktatoren? Neigen beispielsweise freundlichere Personen dazu, dem Empfänger mehr zuzuteilen?
2. **Kognitive Fähigkeiten**: Beeinflussen höhere kognitive Fähigkeiten strategischere oder fairere Entscheidungen?
3. **Studienrichtung**: Gibt es Unterschiede im Entscheidungsverhalten basierend auf dem Studienfach des Subjekts?
4. **Einkommen, Alter und Geschlecht**: Beeinflussen diese demografischen Faktoren die Verteilungsentscheidungen im Diktatorspiel?

### Verknüpfung und Analyse
Durch die Verknüpfung dieser Datensätze können wir beispielsweise untersuchen, ob Personen mit bestimmten Persönlichkeitseigenschaften oder aus bestimmten Studienrichtungen unterschiedliche Entscheidungen im Diktatorspiel treffen. Dies hilft, tiefere Einblicke in die Faktoren zu gewinnen, die das menschliche Entscheidungsverhalten beeinflussen.
